```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import urllib2
import csv
import pandas as pd
import os
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

USERNAME = '14714407056'
PASSWORD = 'hkufina123'
START_PAGE = 26
END_PAGE = 61
driver = webdriver.Firefox()
driver.get("http://simu.eastmoney.com")

df_columns = ['ranking','fund_short_name','fund_company','investment_strategy','lastest_NAV','cumulative_return','recent_month_return','annalized_return','sharpe_ratio','year_rating']
df_etf = pd.DataFrame(columns=df_columns)

def go_to_page(page):
    if page != 1:       
        ipage_indicator = driver.find_element_by_name('p_select')
        ipage_indicator.send_keys(page)
        iconfirm_button = driver.find_element_by_xpath('//*[@id="tab-1-1"]/table/tfoot/tr/td/div[1]/a[5]').click()
        time.sleep(3)
    else:
        return
    
def additional_etfs():
    #uncheck the first checkbox
    icheckbox1 = driver.find_element_by_xpath('//*[@id="ranking-tab-content"]/div[1]/div[3]/div[2]/div[2]').click()
    time.sleep(1)
    #check the rest of the checkbox
    for i in range(7):
        xpath = '//*[@id="ranking-tab-content"]/div[1]/div[3]/div[2]/div[' + str(i+3) +']'
        driver.find_element_by_xpath(xpath).click()
        time.sleep(1)
    
def main():
    driver.implicitly_wait(10)
    #driver.execute_script("return $('#gr-box').remove();");
    #driver.execute_script("return $('#shadow').remove();");
    ilogin_button = driver.find_element_by_xpath('//*[@id="gr-register-step1"]/div[9]/strong').click()
    driver.implicitly_wait(5)
    ilogin_name = driver.find_element_by_id('gr_input_nickname')
    ilogin_name.send_keys(USERNAME)

    ikeyword = driver.find_element_by_id('gr_input_pwd1')
    ikeyword.send_keys(PASSWORD)
    
    isubmit_button = driver.find_element_by_id('gr-login-btn').click()
    time.sleep(3)
    print 'LOGIN SUCCESS'
    
    #go to START PAGE
    additional_etfs()
    go_to_page(START_PAGE)
    print 'Start at PAGE'+ str(START_PAGE)
    
    n = 1
    for CURRENT_PAGE in range(START_PAGE,END_PAGE+1):
        print '--------------------PAGE ' + str(CURRENT_PAGE) +' START--------------------'
        START_TIMESTAMP = time.time()
        for row in driver.find_elements_by_xpath('//*[@id="tab-1-1"]/table/tbody/tr'):
            print 'Start handling '+ str(n) +'th row'
            for i in range(10):
                if df_columns[i] == 'year_rating':
                    count = len(row.find_elements_by_xpath('td/div/i[contains(@class, "icon-rank-red-star")]'))
                    if count > 0:
                        df_etf.loc[n, df_columns[i]] = str(count) + 'out of 5'
                    else:
                        df_etf.loc[n, df_columns[i]] = 'Not provided'
                else:
                    df_etf.loc[n, df_columns[i]] = row.find_elements_by_tag_name("td")[i].text
            print 'Finish handling '+ str(n) +'th row'
            print '----------------------------------'
            n = n + 1
        print '--------------------PAGE ' + str(CURRENT_PAGE) +' FINISH--------------------'
        END_TIMESTAMP = time.time()
        print '-TIME USED:' + str(START_TIMESTAMP-END_TIMESTAMP) + '-'
        
        #store every first 5,10,15... pages to CSV for backup purpose
        if (CURRENT_PAGE % 5) == 0:
            filename = '[Bonus][Intrim Backup][' + str(CURRENT_PAGE) + ']Project 2.1.2 PAGE ' + str(START_PAGE) + 'to PAGE ' + str(END_PAGE) + 'Records.csv'
            df_etf.to_csv(filename, index=False,  encoding='utf-8')
            print '[Intrim Backup]Saved First '+ str(CURRENT_PAGE) + 'Pages Records to CSV'
    
        #click next page
        driver.find_element_by_xpath('//*[@id="tab-1-1"]/table/tfoot/tr/td/div[1]/a[3]').click()
        time.sleep(3)
    
    filename = '[Bonus]Project 2.1.2 PAGE ' + str(START_PAGE) + 'to PAGE ' + str(END_PAGE) + 'Records.csv'
    df_etf.to_csv(filename, index=False, encoding='utf-8')
    print 'Exported to csv'
    driver.close()
if __name__ == "__main__":
    main()

```
