```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import urllib2
import csv
import pandas as pd
import os
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

START_PAGE = 6
END_PAGE = 10
driver = webdriver.Firefox()
driver.get("http://fund.jrj.com.cn/funddata/smjj/")

df_columns = ['ranking','fund_short_name','fund_company','investment_strategy','lastest_NAV','cumulative_return','recent_month_return','annalized_return','sharpe_ratio','year_rating']
df_etf = pd.DataFrame(columns=['Ranking','Fund Short Name','Fund Company','Dummy','Lastest NAV','Dummy','Recent Month Return','Dummy','Annalized Return','Dummy','Cumulative Return','Sharpe Ratio','Year Rating','Investment Strategy'])

def go_to_page(page):
    if page != 1:
        for i in range(page-1):
            driver.find_element_by_xpath('//*[@id="pagefoot"]/a[text()="下一页>>"]').click()
            time.sleep(1)
    else:
        return

def main():
    time.sleep(3)
    #change page size to 50
    driver.find_element_by_xpath('//*[@id="ps"]/a[3]').click()
    time.sleep(3)
    go_to_page(START_PAGE)  
    time.sleep(1)
    
    print 'Start at PAGE'+ str(START_PAGE)
    
    n = 1
    for CURRENT_PAGE in range(START_PAGE,END_PAGE+1):
        print '--------------------PAGE ' + str(CURRENT_PAGE) +' START--------------------'
        START_TIMESTAMP = time.time()
        for row in driver.find_elements_by_xpath('//*[@id="table_yield"]/tbody/tr'):
            for i in range(11):
                df_etf.loc[n, df_etf.columns[i]] = row.find_elements_by_tag_name("td")[i].text
            n = n + 1
        print '--------------------PAGE ' + str(CURRENT_PAGE) +' FINISH--------------------'
        END_TIMESTAMP = time.time()
        print '-TIME USED:' + str(START_TIMESTAMP-END_TIMESTAMP) + '-'
        
        #store every first 5,10,15... pages to CSV for backup purpose
        if (CURRENT_PAGE % 10) == 0:
            filename = '[Bonus][jrj][Intrim Backup][' + str(CURRENT_PAGE) + ']Project 2.1.2 PAGE ' + str(START_PAGE) + 'to PAGE ' + str(CURRENT_PAGE) + 'Records.csv'
            df_etf.to_csv(filename, index=False,  encoding='utf-8')
            print '[Intrim Backup]Saved First '+ str(CURRENT_PAGE) + 'Pages Records to CSV'
    
        #click next page
        driver.find_element_by_xpath('//*[@id="pagefoot"]/a[text()="下一页>>"]').click()
        time.sleep(2)
    
    filename = '[Bonus][jrj]Project 2.1.2 PAGE ' + str(START_PAGE) + 'to PAGE ' + str(END_PAGE) + 'Records.csv'
    df_etf.to_csv(filename, index=False, encoding='utf-8')
    print 'Exported to csv'
    driver.close()
if __name__ == "__main__":
    main()

```
