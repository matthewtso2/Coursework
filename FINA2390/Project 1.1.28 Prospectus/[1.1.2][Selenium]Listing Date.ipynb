```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import urllib2
import csv
import pandas as pd
import os
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

START_PAGE = 6
END_PAGE = 10
driver = webdriver.Firefox()
df = pd.DataFrame.from_csv('Stock code.csv', index_col=None)
df['Listing Date'] = 0

def build_url(stock_code):
    url = "http://www.hkex.com.hk/Market-Data/Securities-Prices/Equities/Equities-Quote?sym="+ str(stock_code) +"&sc_lang=en"
    return url

def main():    
    for i in range(len(df)):
        driver.get(build_url(df.loc[i, 'Stock Code']))
        time.sleep(1.5)
        list_date = driver.find_element_by_xpath('//*[@id="lhkexw-quoteequities"]/div[2]/div[3]/div[1]/div/div[2]/div[1]/div[3]/span[2]').text
        print list_date
        df.loc[i, 'Listing Date'] = list_date
        driver.close()
if __name__ == "__main__":
    main()


```

```python
df.to_csv('Stock code with listing date.csv')
```
